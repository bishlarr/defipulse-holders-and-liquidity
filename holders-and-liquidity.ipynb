```python
def scrap_holders_by_address(contract_address):

  BASE_URL = "https://etherscan.io/token/{}".format(contract_address)
  resp = scraper.get(BASE_URL)
  holders = re.search(r'number of holders (.*?) and', resp.text).group(1).replace(',', '')
  #name = re.search(r'\| (.*?) \(', resp.text).group(1)
  #symbol = re.search(r'\((.*?)\)', resp.text).group(1)

  return holders

def interactive_display(df):

  data_table.enable_dataframe_formatter()
  display(df)
  data_table.disable_dataframe_formatter()
  
  return
```
